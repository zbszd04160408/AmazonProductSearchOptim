## 1. Check if the search query is included in the title, description, bullet point, brand or color. 

```python
def is_included_in_title(row):
  return int(str(row['query']).lower() in str(row['product_title']).lower())
    
def is_included_in_description(row):
  return int(str(row['query']).lower() in str(row['product_description']).lower())

def is_included_in_bullet_point(row):
  return int(str(row['query']).lower() in str(row['product_bullet_point']).lower())

def is_included_in_brand(row):
  return int(str(row['query']).lower() in str(row['product_brand']).lower())
    
def is_included_in_color_name(row):
  return int(str(row['query']).lower() in str(row['product_color_name']).lower())

def is_included_in_any(row):
  return row['is_included_in_title'] or row['is_included_in_description'] or row['is_included_in_bullet_point'] or row['is_included_in_brand'] or row['is_included_in_color_name']
```

```python
merged_df['is_included_in_title'] = merged_df.apply(is_included_in_title, axis=1)
merged_df['is_included_in_description'] = merged_df.apply(is_included_in_description, axis=1)
merged_df['is_included_in_bullet_point'] = merged_df.apply(is_included_in_bullet_point, axis=1)
merged_df['is_included_in_brand'] = merged_df.apply(is_included_in_brand, axis=1)
merged_df['is_included_in_color_name'] = merged_df.apply(is_included_in_color_name, axis=1)

merged_df.head()
```

```python
merged_df['is_included_in_any'] = merged_df.apply(is_included_in_any, axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
! pip install transformers
from transformers import AutoTokenizer, AutoModel, TFAutoModel, pipeline
import numpy as np
from scipy.spatial.distance import cosine
from collections import defaultdict
```

```python
MODEL = 'fabriceyhc/bert-base-uncased-amazon_polarity' #"amazon/bort" #fabriceyhc/bert-base-uncased-amazon_polarity
fill_mask = pipeline("fill-mask", model=MODEL, tokenizer=MODEL)
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModel.from_pretrained(MODEL)
```

```python
def get_embedding(text):
  encoded_input = tokenizer(text, return_tensors='pt')
  features = model(**encoded_input)
  features = features[0].detach().cpu().numpy() 
  features_mean = np.mean(features[0], axis=0) 
  return features_mean

def get_sim(row):
  detail = ""
  if len(row['product_description']) == 0:
    detail = row['product_title']  + row['product_bullet_point'] + row['product_brand']
    detail = detail[:512]
    sim = 1-cosine(get_embedding(row['query']),get_embedding(detail))
  else:
    if len(row['product_description']) > 512:
      row['product_description'] = row['product_description'][:512]
    sim = 1-cosine(get_embedding(row['query']),get_embedding(row['product_description']))
  return sim
```
